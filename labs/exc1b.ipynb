```python

import pandas as pd
data = pd.read_csv("kd.csv", sep=",",header = 0, index_col = None, na_values="?")
df = pd.DataFrame(data)
print(df)

```

```python
affected = df.loc[df['class'] == 'ckd']
print(affected)
res = affected.describe()
print(res)
```

```python
affected.corr()
```

```python
import matplotlib.pyplot as plt
import numpy
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(affected.corr(), vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,11,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(affected.corr().columns.values)
ax.set_yticklabels(affected.corr().columns.values)
plt.show()
```
